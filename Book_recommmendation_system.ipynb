# Popularity based Recommendation System

Our formula is not much complex. We will show only those books which have higher avg rating

```python
#for this we need to build a dataset having books, and its avg rating one one column
ratings_with_name = ratings.merge(books,on='ISBN')
ratings_with_name # for each rating row,(in rating dataset) we also write the name of corresponding book, and its details(so we merge on the column ISBN)
```

```python
# We will consider only those books in the top 50 books which are rated by atleast 250 users(i.e, number of rating of that book is >=250)

#Fisrt we will try to find the number of rating of all books(i.e, how many times they are rated)(We not did group by on ISBN as int this dataset same book have multiple ISBN code

ratings_with_name.groupby('Book-Title').count() #used .count() to find count of it

#see it gives for each book(row) how many people have rated it
#from this we will take Book-Rating,Book-Title
```

```python
ratings_with_name.groupby('Book-Title').count()['Book-Rating']

#to make it a proper dataframe, we will reset index
```

```python
number_of_ratings=ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
number_of_ratings
```

```python
# We will change the column name Book-Rating as number_of_ratings (as its too much confusing)
number_of_ratings.rename(columns={'Book-Rating':'number_of_ratings'},inplace=True)
number_of_ratings
```

```python
# Now we will make another datset which should have name of that book, and the average rating of that book

#By adding numeric_only=True to the mean() function, you instruct pandas to only consider numeric columns when calculating the mean. This will prevent the error caused by attempting to calculate the mean of non-numeric columns.
ratings_with_name.groupby('Book-Title').mean(numeric_only=True)

#from this, we only take Book-rating(which is actually avg rating)
```

```python
avg_rating=ratings_with_name.groupby('Book-Title').mean(numeric_only=True)['Book-Rating'].reset_index()
```

```python
#rename the column Book Rating -> avg rating(to avoid confusion)
avg_rating.rename(columns={'Book-Rating':'avg_rating'},inplace=True)
avg_rating
```

```python
# Now number_of_rating + avg_rating combined
popular_df=number_of_ratings.merge(avg_rating,on='Book-Title')
popular_df
```

```python
# Now we have a dataset having only the name of book, no. of times its rated and its avg rating
# So now we will select only those books(i.e, rows) which have rated >=250 times

popular_df = popular_df[popular_df['number_of_ratings']>=250]

```

```python
popular_df # See how number of books decreased
```

```python
# Now we sort them in decreasing order of their avg rating
popular_df=popular_df.sort_values('avg_rating',ascending=False)
```

```python
# now we want only top 50 books
popular_df=popular_df[0:50]
popular_df
```

```python
#Now for all these books i want to display these top 50 books by not just name, but also its details(picture, author name,publisher name etc)
popular_df=popular_df.merge(books,on='Book-Title')
```

```python
popular_df
```

```python
# Now when we are merging popular_df on the books df, then due to having duplicate ISBN numbers for same book, we get >50 rows in our dataset
#so we droop duplicate books
popular_df=popular_df.drop_duplicates('Book-Title')
```

# Collaborative Filtering based Recommendation System

```python
ratings_with_name
#from this data frame, we will take only those users who have given rating >=200
#why?--->because then this means that we will be cosidering rating of some people which are genuinely passionate of books, and they keep on reading books and rate them
#    --->so there rating might depict true picture of the book. Thus our model will be trained on much better data
```

```python
ratings_with_name.groupby('User-ID').count()
#so we got user id and the number of rating it did to the books
#we will take only those ones which have done rating >=200
```

```python
ratings_with_name.groupby('User-ID').count()['Book-Rating']>=200
```

```python
x=ratings_with_name.groupby('User-ID').count()['Book-Rating']>=200
```

```python
x[x] #-->doing boolean indexing
# we will store this in the dataframe
```

```python
good_users=x[x].index
good_users# it contains all those users
```

```python
#now from rating_with_name, we will filter suhc that only those users exists which have given >=200 ratings
filtered_rating=ratings_with_name[ratings_with_name['User-ID'].isin(good_users)]
```

```python
filtered_rating
```

```python
# in this same way we will filter such that only those books exists which are given review by >=50 people
filtered_rating.groupby('Book-Title').count()
```

```python
y=filtered_rating.groupby('Book-Title').count()['Book-Author']>=50 # getting a boolean series
y[y] # doing boolean indexing
```

```python
famous_books = y[y].index
famous_books
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_ratings
```

```python
# making a pivot table
# rows mei books, columns mei userID and each cell will have rating of book
pt = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt
```

```python
# many NaN values so we place it with 0 rating(its NaN as they haven't rated)
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
# Now next step of collaborative filtering is finding eucliead distance between all points

# each book is imagined like a vector <user1'rating, user2'rating, user3'rating.......> in 815 dimentional space(the more the numbe rof good users, the higher dimensional space in)
# (we imagine one book as one point in 815 dimensional coordinate system, ad we find closer point closer to one point

#so, for each vector, we find out all the near vector (so for this we find out distance from each vector (i.e, book))
#as by the help of distance we would be able to tell how similar this book is
#1) finding euclidead distance between all pair of vector(i.e, books)

#for each row of pivot table(ie, for each book), we find cosine similarity between each row
from sklearn.metrics.pairwise import cosine_similarity
similarity_scores = cosine_similarity(pt)

```

```python
similarity_scores
```

```python
similarity_scores[0]
```

```python
similarity_scores.shape # for each books, we have got similarity score between each 707 book(including itself)
```

```python
# we make this recommend function , in which we will give one book, and it will give 5 similar books

def recommend(book_name):
  #first we need to find in which index this book is present in the pivot table
  index = np.where(pt.index==book_name)[0][0] #index fetching
  #then will go to that index(i.e, row of similarity_index)
  #on enumerating we will get the book index and its similarity score
  #then we will sort it in decreasing order of the the similarity score(i.e, second value)
  # then as we want to show top 5 similar books(except that one only) we skip the first one(as it will obviosly have the highest similar score of 1)
  similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:6]

  #this willl contain the indexes of the top 5 similar books

  for i in similar_items:
    # to get the name of book, we used pivot table
    print(pt.index[i[0]])
```

Lets test this

```python
recommend('1984')
```

```python
recommend('Message in a Bottle')
# its a romantic movie, so yes all other should be similar(and they are)
```

```python
recommend('The Fellowship of the Ring (The Lord of the Rings, Part 1)')
```

The recommend function is going fine, but we are getting only the name of the book. But in wesite when we will recommend it we would love to have the name of book, author, their rating, and their images

So now we will edit it accordingly

```python
def Recommend(book_name):
    # index fetch
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:5]

    data = []
    for i in similar_items:
        item = [] #all data iten that we want for one book
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))

        data.append(item) # for each book, appending all of its qualitites

    return data
```

```python
Recommend('The Fellowship of the Ring (The Lord of the Rings, Part 1)')
```
